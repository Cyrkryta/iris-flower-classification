# Data collection and Analysis

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import scipy.stats as st
```

```python
iris_data = pd.read_csv("Iris.csv") # read csv file
df = pd.DataFrame(iris_data)        # Creating dataframe
print(iris_data.shape)              # shape
print(iris_data.describe())         # description
print(iris_data.columns)            # columns
print(iris_data.dtypes)             # types
print(iris_data.head())             # data visuals

# Update dataset names
iris_data = iris_data.rename(columns={'SepalLengthCm':'sepal_l', 'SepalWidthCm':'sepal_w', 'PetalLengthCm':'petal_l', 'PetalWidthCm':'petal_w', 'Species':'species'})
iris_data.head()
```

```python
# Visualize the data
# Difference between sepal and petal length
sns.lineplot(x=iris_data['sepal_l'], y=iris_data['petal_l'])
sns.scatterplot(x=iris_data['sepal_l'],y=iris_data['petal_l'])
plt.title('Sepal length VS. Petal length')
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
```

### Comment on plot
From the plot we can see that the petal are generally smaller than the sepal length. 

```python
# Multivariate analysis
sns.pairplot(iris_data, hue='species', palette='CMRmap_r')
```
